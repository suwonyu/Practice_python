```python

%I : 12시간으로 환산한 시간
%M : 분
%S : 초
%p : AM,PM
%

%A : 요일 full
%a : 요일 약자
%w : 숫자요일	0:일요일, 1:월요일 
%j : 1월1일 부터 누적된 날짜(1~366)
%c : 날짜, 시간
%x : 월/일/년도
%X : 시:분:초
%W : 누적된 주(월요일을 시작으로 본)
%U : 누적된 주(일요일을 시작으로 본)
%z : 시간대(대한민국 표준시)	+0900

```

```python
[문제88] 이름, 입사한 요일(한글)을 출력하세요.
```

```python
import csv
import time

file = open('c:\python\emp1.csv','r')
emp_list = csv.reader(file)

def wday(arg):
    x = time.strptime(arg,'%Y-%m-%d')
    d = ['일요일','화요일','수요일','목요일','금요일','토요일','월요일']
    x1 = time.strftime('%w',x)
    return d[int(x1)]

[print(i[1],wday(i[4])) for i in emp_list]
```

```python
#datetime 모듈 사용 
import datetime

def K_wdays(arg): 
        date=datetime.datetime.strptime(arg,"%y/%m/%d")
        date=date.weekday()
        d=["월요일","화요일","수요일","목요일","금요일","토요일","일요일"]
        return d[date]
   
file = open("c:\R\emp.csv","r")
emp_csv = csv.reader(file)
for i in emp_csv:
       print(i[2],K_wdays(i[5]))
```

```python
[문제89] 이름, 입사한 날짜부터 오늘까지 총 몇일 근무했는지 출력하세요.
```

```python
import time
import csv
import math

file = open('c:\python\emp1.csv','r')
emp_csv = csv.reader(file)
for emp_list in emp_csv:
    a = time.mktime(time.localtime())
    b = time.mktime(time.strptime(emp_list[4],"%Y-%m-%d"))
    c = a - b
    print (emp_list[1], math.trunc(c/86400))  #하루 = 86400초
```
