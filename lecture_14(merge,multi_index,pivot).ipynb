```python
#csv
import csv

file = open("c:\python\emp1.csv","r")
emp_csv = csv.reader(file)

cn_null = 0
cn_notn = 0

for i in emp_csv:
    if i[6] == '':
        cn_null += 1
    elif i[6] != '':
        cn_notn += 1
print('null = {}, notnull = {}'.format(cn_null, cn_notn))
print('null = %s, notnull = %s'%(cn_null,cn_notn))
```

```python
file_emp = open("c:\python\emp1.csv","r")
file_dept = open("c:\python\dept.csv","r")

emp_csv = csv.reader(file_emp)
dept_csv = csv.reader(file_dept)

emp=[]
for i in emp_csv:
    emp.append({'empno':i[0],'ename':i[1],'job':i[2],'mgr':i[3],'hire_date':i[4],'sal':i[5],'comm':i[6],'deptno':i[7]})
    
dept=[]
for i in dept_csv:
    dept.append({'deptno':i[0],'dname':i[1],'mgr':i[2],'loc':i[3]})
```

```python
def join_csv(agg1,agg11,agg2,agg22):
    for i in agg1:
        for n in agg2:
            if i[agg11] == n[agg22]:
                print(i,n)
```

```python
join_csv(emp,'deptno', dept,'deptno')
```

```python
join_csv(emp,'mgr', emp,'empno')
```

```python
[문제162] 사원번호를 입력하면 부서이름을 리턴해주는 함수를 생성하세요.
```

```python
file_emp = open("c:\python\emp1.csv","r")
file_dept = open("c:\python\dept.csv","r")

emp_csv = csv.reader(file_emp)
dept_csv = csv.reader(file_dept)

emp=[]
for i in emp_csv:
    emp.append({'empid':i[0],'ename':i[1],'job':i[2],'mgr':i[3],'hire_date':i[4],'sal':i[5],'comm':i[6],'deptno':i[7]})
    
dept=[]
for i in dept_csv:
    dept.append({'deptno':i[0],'dname':i[1],'mgr':i[2],'loc':i[3]})
```

```python
def dname(agg):
    for i in emp:
        if int(i['empid']) == agg:
            for n in dept:
                if i['deptno'] == n['deptno']:
                    print('deptno : {}, dname : {}'.format(i['deptno'],n['dname']))
```

```python
import pandas as pd

emp = pd.read_csv("c:\python\emp1.csv", names = ['empid','name','job','mgr','hire_date','sal','comm','deptno'])
dept = pd.read_csv("c:\python\dept.csv", names = ['deptno','dname','mgr','loc'])
```

```python
pd.merge(emp,dept,on = 'deptno') # 조인 키컬럼 을 지하여 두개의 판다스를 조인(컬럼이름이 동일할 경우) 이퀴조인
```

```python
pd.merge(emp,dept,left_on='deptno',right_on='deptno') # 키컬럼이 다를경우
```

```python
pd.merge(emp,dept,on='deptno',how='left') # 왼쪽 데이터는 다뿌림   left outer join
```

```python
pd.merge(emp,dept,on='deptno',how='right')# 오른쪽 데이터는 다뿌림  right outer join
```

```python
pd.merge(emp,dept,on='deptno',how='outer') # 모든 데이터 다 뿌림    outer join(key값 일치 안되는 것까지)
```

```python
pd.merge(emp,dept,on='deptno',how='inner') # 기본값   inner join(key값 일치되는 것만)
```

```python
# self join
pd.merge(emp, emp, left_on = 'mgr', right_on = 'empid')[['name_x','name_y']]
```
