```python
[문제136] 급여를 많이 받는 순으로 10위 까지를 구하세요. 
```

```python
import pandas as pd
from pandas import Series,DataFrame

emp = pd.read_csv("c:\python\emp1.csv",
                   names=["empid","name","job","mgr","hire_date","sal","comm","deptno"])
```

```python
ㅁapply : 행,열값을 인수값으로 받아서 반복하여 그함수를 적용한다.
```

```python
s1 = Series([1,2,3])
s1
```

```python
s1**2
```

```python
def square(x):
    return x**2
```

```python
square(2)
```

```python
#apply 함수는 행,열값을 인수값으로 받아서 반복하여 그함수를 적용한다.
	#apply함수는 인수값으로 함수가 들어가야 한다.
```

```python
s1.apply(square)
```

```python
#함수대신 람다를 사용
s1.apply(lambda x : x**2)
```

```python
[문제141] 이름 첫글자가 S 로 시작하는 사원들의 이름을 출력하세요.
```

```python
emp = pd.read_csv("c:\python\emp1.csv",
                names=["empid","name","job","mgr","hire_date","sal","comm","deptno"])
#함수 사용
def ini(arg):
    return arg[0]
```

```python
emp.loc[emp['name'].apply(ini)=='S',['name']]
```

```python
#lambda 사용
emp.loc[emp['name'].apply(lambda x: x[0])=='S',['name']]
```
