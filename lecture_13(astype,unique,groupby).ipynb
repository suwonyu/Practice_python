```python

#[선생님]

from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import operator
import math

emp = pd.read_csv("c:\python\emp1.csv", names = ['empid','name','job','mgr','hire_date','sal','comm','deptno'])

deptno_unique = Series(emp['deptno'].unique()) #numpy.array형식이됨
deptno_unique = deptno_unique.dropna()

deptno_dict = {}

for i in deptno_unique:
    dept_sum = emp.loc[emp['deptno'] == i,'sal'].sum()
    deptno_dict[i] = dept_sum

for k,v in sorted(deptno_dict.items(), key=operator.itemgetter(0)):
    print(math.trunc(k),v)
```

```python
file = open("c:\python\emp1.csv","r")
emp_csv = csv.reader(file)

dept_dic={}
year=0
for i in emp_csv:
    year=int(time.strptime(i[4],'%Y-%m-%d').tm_year)
    if year in dept_dic:
        dept_dic[year] = dept_dic[year]+1
    else:
        dept_dic[year] = 1

dept_dic
```
